## 1. Y-axis (Autocorrelation): This shows how much the time series data is correlated with its own previous values at different time lags. 
The values range from -1 to 1. 

A value close to 1 means strong positive correlation (the data points are similar).
A value close to -1 means strong negative correlation (the data points are opposite).
A value close to 0 means no correlation (random relationship).

## X-axis (Lag): This represents the time lag.
A lag of 1 means the correlation between each data point and the one immediately before it, a lag of 2 means the correlation with the data point two steps before, and so on.

```python
# Define the frequency columns
frequency_columns = [
    'freq_1 freq', 'freq_p3_1', 'freq_p4_1', 'freq_p4_2', 
    'freq_p5_1', 'freq_p5_2', 'freq_p6_1', 'freq_p7_1',
    'freq_p7_2', 'freq_p8_1', 'freq_p8_2', 'freq_p9_1',
    'freq_p10_1', 'freq_p11_1', 'freq_p12_1', 'freq_p14_1',
    'freq_p15_1', 'freq_p18_1'
]

# Initialize an empty DataFrame to store cross-correlation coefficients
cross_corr_matrix = pd.DataFrame(index=frequency_columns, columns=frequency_columns)

# Compute cross-correlation for each pair of columns
for col1 in frequency_columns:
    for col2 in frequency_columns:
        if col1 != col2:
            # Compute cross-correlation and find the maximum value
            cross_corr = np.correlate(df[col1] - np.mean(df[col1]), df[col2] - np.mean(df[col2]), mode='full')
            max_corr = np.max(np.abs(cross_corr)) / (np.std(df[col1]) * np.std(df[col2]) * len(df[col1]))
            cross_corr_matrix.loc[col1, col2] = max_corr
        else:
            cross_corr_matrix.loc[col1, col2] = 1  # Correlation with itself

# Convert the matrix values to float for plotting
cross_corr_matrix = cross_corr_matrix.astype(float)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cross_corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Cross-Correlation Matrix of Frequency Columns')
plt.show()
```

## Cross-Correlation Matrix: The cross_corr_matrix DataFrame stores the maximum cross-correlation values between each pair of frequency columns.
Heatmap: The heatmap is a visual representation where each cell's color intensity reflects the strength of the cross-correlation between two columns.
Diagonal Values: The diagonal of the matrix represents the correlation of each frequency with itself, which is always 1.
Insights:
High Cross-Correlation: Darker or more intense colors indicate a higher cross-correlation between two frequency columns, suggesting that these two signals have a strong relationship.
Low Cross-Correlation: Lighter colors indicate weaker or no significant relationship between the two signals.
Customization:
Lag Analysis: If you're interested in specific lags, you can modify the code to store the cross-correlation values for those particular lags.
Thresholding: You can filter the matrix to show only cross-correlations above a certain threshold if you are interested in significant correlations only.

```python
import numpy as np
import pandas as pd
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean
import matplotlib.pyplot as plt

# Your specified frequency columns
frequency_columns = [
    'freq_1 freq', 'freq_p3_1', 'freq_p4_1', 'freq_p4_2', 
    'freq_p5_1', 'freq_p5_2', 'freq_p6_1', 'freq_p7_1',
    'freq_p7_2', 'freq_p8_1', 'freq_p8_2', 'freq_p9_1',
    'freq_p10_1', 'freq_p11_1', 'freq_p12_1', 'freq_p14_1',
    'freq_p15_1', 'freq_p18_1'
]

# Initialize a dictionary to store DTW distances
dtw_distances = {}

# Calculate DTW distances between all pairs of frequency columns
for i in range(len(frequency_columns)):
    for j in range(i + 1, len(frequency_columns)):
        ts1 = df[frequency_columns[i]].dropna().values
        ts2 = df[frequency_columns[j]].dropna().values

        # Ensure that the data is 1-D by flattening it
        ts1 = np.ravel(ts1)
        ts2 = np.ravel(ts2)

        # Check if ts1 and ts2 are non-empty and 1-D
        if ts1.ndim != 1 or ts2.ndim != 1:
            print(f"Skipping {frequency_columns[i]} and {frequency_columns[j]} due to dimensionality issues")
            continue

        # Ensure ts1 and ts2 are not empty
        if len(ts1) == 0 or len(ts2) == 0:
            print(f"Skipping {frequency_columns[i]} and {frequency_columns[j]} due to insufficient data")
            continue

        # Calculate DTW distance
        distance, path = fastdtw(ts1, ts2, dist=euclidean)
        dtw_distances[(frequency_columns[i], frequency_columns[j])] = distance

        print(f'DTW Distance between {frequency_columns[i]} and {frequency_columns[j]}: {distance}')

# Visualize the alignment between the first pair of frequency columns as an example
if dtw_distances:
    first_pair = list(dtw_distances.keys())[0]
    ts1 = df[first_pair[0]].dropna().values
    ts2 = df[first_pair[1]].dropna().values

    # Ensure the data is 1-D by flattening it
    ts1 = np.ravel(ts1)
    ts2 = np.ravel(ts2)

    _, path = fastdtw(ts1, ts2, dist=euclidean)

    plt.figure(figsize=(10, 6))
    plt.plot(ts1, label=first_pair[0], color='blue')
    plt.plot(ts2, label=first_pair[1], color='red')

    # Add lines to indicate the alignment path
    for (i, j) in path:
        plt.plot([i, j], [ts1[i], ts2[j]], color='gray', linestyle='--')

    plt.title(f'Dynamic Time Warping Alignment: {first_pair[0]} vs {first_pair[1]}')
    plt.legend()
    plt.show()
else:
    print("No valid DTW comparisons could be made.")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[frequency_columns])

# Apply PCA
pca = PCA(n_components=2)  # Choosing 2 components for visualization
principal_components = pca.fit_transform(scaled_data)

# Plotting the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1], alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Frequency Columns')
plt.grid(True)
plt.show()

# Explained variance
print(f"Explained variance by principal components: {pca.explained_variance_ratio_}")
```

## 2. Principal Component Analysis (PCA):
Objective: To reduce the dimensionality of the data while retaining the most important variance. This helps in identifying common underlying trends.
Method: Use PCA to project the frequency columns onto fewer dimensions (principal components) that capture the majority of the variance

Interpretation: Clusters in the PCA plot indicate common trends among the frequency columns. Components that explain high variance point to significant patterns.

```python
from tslearn.clustering import TimeSeriesKMeans
from sklearn.preprocessing import MinMaxScaler

# Scale data for clustering
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[frequency_columns])

# Apply TimeSeriesKMeans clustering
model = TimeSeriesKMeans(n_clusters=3, metric="euclidean", random_state=0)
clusters = model.fit_predict(scaled_data)

# Add cluster labels to the dataframe
df['Cluster'] = clusters

# Visualize clusters
plt.figure(figsize=(10, 6))
for cluster_num in np.unique(clusters):
    cluster_data = df[df['Cluster'] == cluster_num][frequency_columns].T
    plt.plot(cluster_data, alpha=0.5, label=f'Cluster {cluster_num}')

plt.title('Clustering of Frequency Columns')
plt.xlabel('Time')
plt.ylabel('Frequency Values')
plt.legend()
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated time series data to represent freq_p3_1 as we don't have raw data
# Let's assume a periodic signal with some noise for illustration
np.random.seed(0)
t = np.linspace(0, 100, 1000)  # 1000 time points
# Simulating a time series with two frequencies: 1/20 and 1/50 and some random noise
signal = 3 * np.sin(2 * np.pi * t / 20) + 2 * np.sin(2 * np.pi * t / 50) + np.random.normal(0, 0.5, t.size)

# Apply Fourier Transform
fourier_transform = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(t), t[1] - t[0])

# We take the absolute value of the Fourier coefficients (magnitude)
magnitude = np.abs(fourier_transform)

# Plot the frequency spectrum
plt.figure(figsize=(10, 6))
plt.plot(frequencies[:len(frequencies)//2], magnitude[:len(frequencies)//2])  # Only plot positive frequencies
plt.title('Frequency Spectrum of freq_p3_1 Time Series')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

Explanation of the Fourier Transform Results:
The frequency spectrum above shows how much each frequency contributes to the original time series. Here's a breakdown of what we're seeing:

Dominant Low Frequencies: The largest peak is at the lowest frequency range, indicating that the time series has a strong low-frequency (longer-term) component. This suggests there is some slow, gradual oscillation in the data.

No Significant High Frequencies: The rest of the frequencies have much lower magnitudes, indicating that there aren't strong high-frequency (short-term) oscillations.

Interpretation: This means the oscillations in the autocorrelation of freq_p3_1 are largely driven by slow, recurring trends. There might be cycles in the data with a period related to this dominant low frequency.
