```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dictionary to store DataFrames and filenames
data_dict = {}
for filename in ["dict_new_1.csv", "dict_new_p3_1.csv", "dict_new_p4_1.csv", "dict_new_p5_1.csv", "dict_new_p6_1.csv"]:
    data_dict[filename] = pd.read_csv(f"./dict_spandan/{filename}")  # Assuming file paths are relative

plt.figure(figsize=(16, 8))

# Iterate over the dictionary
for filename, df in data_dict.items():
    # Ensure consistent column handling (assuming first column is 'Time_Frame')
    if 'Time_Frame' in df.columns:
        time_col = 'Time_Frame'
    elif 'Time Frame' in df.columns:  # Handle potential alternative column name
        time_col = 'Time Frame'
    else:
        raise ValueError("First column name for time data is not 'Time_Frame' or 'Time Frame'")

    # Extract frequency columns (assuming first column is time data)
    frequency_cols = df.columns[1:]

    # Plot each frequency column with its corresponding CSV filename in the label
    for col in frequency_cols:
        plt.plot(df[time_col], df[col], label=f"{filename} - {col}")  # Customize label format

# ... (rest of the code for title, labels, and legend remains the same)


# Adding title and labels
plt.title('Frequency Trends Over Time')
plt.xlabel('Time Frame')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')

# Adding legend to show each line represents which frequency column
plt.legend(title='Frequency Columns', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'files' is a list containing paths to your CSV files

plt.figure(figsize=(16, 8))

# Iterate over each CSV file
for i, file in enumerate(files):
    # Load data from the CSV file
    df = pd.read_csv(file)

    # Ensure consistent column handling (assuming first column is 'Time_Frame')
    if 'Time_Frame' in df.columns:
        time_col = 'Time_Frame'
    elif 'Time Frame' in df.columns:  # Handle potential alternative column name
        time_col = 'Time Frame'
    else:
        raise ValueError("First column name for time data is not 'Time_Frame' or 'Time Frame'")

    # Extract frequency columns (assuming first column is time data)
    frequency_cols = df.columns[1:]

    # Plot each frequency column with its corresponding CSV filename in the label
    for col in frequency_cols:
        plt.plot(df[time_col], df[col], label=f"{file.split('/')[-1]} - {col}")  # Customize label format

# Add title and labels
plt.title('Frequency Trends Over Time')
plt.xlabel('Time Frame')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')

# Show the legend with filenames
plt.legend(title='Frequency Columns by CSV File', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
