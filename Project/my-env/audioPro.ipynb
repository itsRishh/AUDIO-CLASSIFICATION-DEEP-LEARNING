```python
import pandas as pd
import numpy as np
from scipy.fftpack import fft

# Load data from CSV
data = pd.read_csv('../dict_spandan/dict_new_1.csv', header=None, skiprows=1)

# Extract time stamps and frequencies
time_stamps = data[0].values
frequencies = data[1].values

# Calculate ZCR
zcr = np.sum(np.diff(np.sign(frequencies)) != 0) / len(frequencies)

# Calculate RMS
rms = np.sqrt(np.mean(frequencies**2))

# Calculate FFT
fft_result = fft(frequencies)

# Calculate spectral centroid
spectral_centroid = np.sum(np.arange(len(fft_result)) * np.abs(fft_result)**2) / np.sum(np.abs(fft_result)**2)

print("zcr : ", zcr, "\n","fft_result : ",fft_result, "\n","rms : ",rms, "\n","spectral_centroid : ",spectral_centroid, "\n", )

# ... (other features)
```
